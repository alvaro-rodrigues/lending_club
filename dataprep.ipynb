```python
df = pd.read_csv('new_loan.csv', low_memory=False)
```

```python
df = df[df.loan_status != 'Current']
```

```python
df_null = pd.DataFrame({'Count': df.isnull().sum(), 'Percent': 100*df.isnull().sum()/len(df)})
df_null[df_null['Count'] > 0]
```

```python
df = df.dropna(axis=1, thresh=int(0.80*len(df)))
```

```python
df.info()
```

```python
num_feat = df.select_dtypes('number').columns.values
df[num_feat].nunique().sort_values()
```

```python
df = df.drop(['policy_code'], axis=1, errors='ignore')
```

```python
#correlacao entre adrr_state e purpose
df = df.drop(['title', 'zip_code'], axis=1, errors='ignore')
```

```python
df_null = pd.DataFrame({'Count': df.isnull().sum(), 'Percent': 100*df.isnull().sum()/len(df)})
df_null[df_null['Count'] > 0]
```

```python
fill_empty = ['emp_title']

fill_max = ['bc_open_to_buy', 'mo_sin_old_il_acct', 'mths_since_recent_bc', 'mths_since_recent_inq',
            'pct_tl_nvr_dlq', 'mo_sin_old_rev_tl_op', 'mo_sin_rcnt_rev_tl_op', 'mo_sin_rcnt_tl']

fill_zero = ['emp_length', 'delinq_2yrs', 'inq_last_6mths', 'pub_rec', 'total_acc', 'annual_inc', 'dti',
            'tot_coll_amt', 'total_rev_hi_lim', 'acc_open_past_24mths', 'avg_cur_bal', 'tot_cur_bal',
            'bc_util', 'chargeoff_within_12_mths', 'delinq_amnt', 'mort_acc', 'num_accts_ever_120_pd',
            'num_actv_bc_tl', 'num_actv_rev_tl', 'num_rev_accts', 'num_bc_tl', 'num_il_tl', 'num_op_rev_tl',
            'num_rev_accts', 'num_rev_tl_bal_gt_0', 'num_sats', 'num_tl_120dpd_2m', 'num_tl_30dpd',
            'num_tl_90g_dpd_24m', 'num_tl_op_past_12m', 'pub_rec_bankruptcies', 'tax_liens', 'tot_hi_cred_lim',
            'total_bal_ex_mort', 'total_bc_limit', 'total_il_high_credit_limit', 'acc_now_delinq', 'num_bc_sats']

fill_mean = ['open_acc', 'revol_util', 'collections_12_mths_ex_med', 'percent_bc_gt_75']

df[fill_empty] = df[fill_empty].fillna('')
df[fill_max] = df[fill_max].fillna(df[fill_max].max())
df[fill_zero] = df[fill_zero].fillna(0)
df[fill_mean] = df[fill_mean].fillna(df[fill_mean].mean())
```

```python
df = df.drop(['earliest_cr_line', 'last_pymnt_d', 'last_credit_pull_d'], axis=1, errors='ignore')
```

```python
df_null = pd.DataFrame({'Count': df.isnull().sum(), 'Percent': 100*df.isnull().sum()/len(df)})
df_null[df_null['Count'] > 0]
```
